```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Corona_data/covid_19_clean_complete.csv')
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
df.columns = ['지역', '국가', '위도', '경도', '일자', '확진', '사망', '완치', '치료중', 'WHO']
df.head()
```

## 2-5. EDA(탐색적 데이터분석)

### 1. 대한민국 코로나 현황

```python
K_df = df[df['국가'].str.contains('Korea')]
K_df.head()
```

```python
K_df.reset_index(inplace = True)
K_df.head()
```

```python
plt.figure(figsize = (10, 5))
sns.lineplot(x = '일자', y = '확진', data = K_df)
sns.lineplot(x = '일자', y = '사망', data = K_df)
sns.lineplot(x = '일자', y = '완치', data = K_df)
plt.legend(['확진', '사망', '완치'])
plt.show()
```

### 2. 국가별 코로나 현황 분석

```python
latestDF = df[df['일자'] == max(df['일자'])]
latestDF.head()
```

```python
sortedDF = latestDF.sort_values(by = '확진', ascending = False)
sortedDF.head(10)
```

```python
plt.xticks(rotation = 45)
sns.barplot(x = '국가', y = '확진', data = sortedDF.head(10))
plt.show()
```

```python
# 전세계 확진자 현황
globalDF = df.groupby('일자')[['확진', '사망', '완치']].sum()
globalDF.head()
```

```python
sns.lineplot(data = globalDF)
plt.show()
```
